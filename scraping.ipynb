### Creating Function to return list of appids

```python
# Testing a link in this bottom example to see if I can get the link I need.

#driver = webdriver.Safari()
#driver.get('https://play.google.com/store/apps/collection/cluster?clp=0g4fCh0KF3RvcGdyb3NzaW5nX0dBTUVfQUNUSU9OEAcYAw%3D%3D:S:ANO1ljIOh7o&gsr=CiLSDh8KHQoXdG9wZ3Jvc3NpbmdfR0FNRV9BQ1RJT04QBxgD:S:ANO1ljLWGVQ&hl=en_US')

def scroll_to_bottom(driver):

    old_position = 0
    new_position = None

    while new_position != old_position:
        # Get old scroll position
        old_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
        
        # Sleep and Scroll
        time.sleep(1)
        driver.execute_script((
                "var scrollingElement = (document.scrollingElement ||"
                " document.body);scrollingElement.scrollTop ="
                " scrollingElement.scrollHeight;"))
        
        # Get new position
        new_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))

#scroll_to_bottom(driver)

#html = driver.page_source

#soup = BeautifulSoup(html, 'lxml')

#driver.quit()
```

```python
def cleaner(stuff):
    # Only looking for href links as this is where our appid is.
    links = [a['href'] for a in stuff.find_all('a', href=True)]
    
    # Only looking for unique links as game links show up several times.
    uniquelinks = []
    for x in links:
        if x not in uniquelinks:
            uniquelinks.append(x)
            
    # Only want links with that specific prefix as there are a lot of useless href links.            
    uniqueappidlinks = list(filter(lambda k: '/store/apps/details?id=' in k, uniquelinks))
    
    # Removing the front part now that we have the link we need.
    uniqueappid = [s.replace('/store/apps/details?id=', '') for s in uniqueappidlinks]
    
    return uniqueappid
```

```python
# Putting everything together.

def giveappid(link):
    driver = webdriver.Safari()
    driver.get(link)
    
    scroll_to_bottom(driver)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    driver.quit()
    
    return cleaner(soup)
```

# Getting Reviews for each mobile game

```python
# If a game does not have 250 reviews, it will return as many reviews as possible

from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.nianticlabs.pokemongo',
    lang='en',
    country='us', 
    sort=Sort.MOST_RELEVANT, 
    count=250,
)

result[:4]
```

```python
# turn the above into a function

def givemereviews(appids):
    
    # Parts of the review that we are interested in
    keys = {'content', 'score'}
    
    # The function itself
    result, continuation_token = reviews(
    appids,
    lang='en',
    country='us', 
    sort=Sort.MOST_RELEVANT, 
    count=250,
)
    # Filtering the results
    filteredresults = [{k:v for k, v in i.items() if k in keys} for i in result]
    for item in filteredresults:
        item.update( {"appId":appids})
    return filteredresults
```

```python
# It takes about 1 minute to run 100 games

starttime = datetime.now()

testlist = [item for sublist in list(map(lambda appid: givemereviews(appid), uniqueappidlist[:100])) for item in sublist]

print(datetime.now() - starttime)
print('------------')
print(testlist[:5])
```

```python
# We will use bz2 to compress pickle files to 1/4th the size, which brings us just under the Github limit.

import bz2

# This is to compress it.
def compressed_pickle(title, data):
    with bz2.BZ2File(title + '.pbz2', 'w') as f:
        pickle.dump(data, f)
        
# This is to bring it back.        
def decompress_pickle(file):
    data = bz2.BZ2File(file, 'rb')
    data = pickle.load(data)
    return data
```
